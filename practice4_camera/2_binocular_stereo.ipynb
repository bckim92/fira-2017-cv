```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

imgL = cv2.imread('images/aloeL.jpg', 0)
imgR = cv2.imread('images/aloeR.jpg', 0)
imgGT = cv2.imread('images/aloeGT.png', 0)

stereo = cv2.StereoBM_create(numDisparities=96, blockSize=15)
disparity = stereo.compute(imgL,imgR)

plt.figure(figsize=(16,12))
plt.subplot(2,2,1); plt.title('Left image'); plt.imshow(imgL, cmap='gray'); plt.axis('off')
plt.subplot(2,2,2); plt.title('Right image'); plt.imshow(imgR, cmap='gray'); plt.axis('off')
plt.subplot(2,2,3); plt.title('Estimated disparity'); plt.imshow(disparity, cmap='gray'); plt.axis('off')
plt.subplot(2,2,4); plt.title('Ground truth'); plt.imshow(imgGT, cmap='gray'); plt.axis('off')
```
